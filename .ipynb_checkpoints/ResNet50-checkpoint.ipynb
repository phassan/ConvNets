## ResNet 50

### Deep Residual Learning for Image Recognition
https://arxiv.org/pdf/1512.03385.pdf

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Add, Dense, ReLU, ZeroPadding2D, BatchNormalization, Flatten, Conv2D, AveragePooling2D, MaxPool2D
from tensorflow.keras.initializers import glorot_uniform
from tensorflow.keras.utils import plot_model
```

### Identity Block

<img src="images/Identity-Block.png" style="width=45px;height=45"/>

```python
def identity_block(x, ks, filters):

    f1, f2, f3 = filters
    x_shortcut = x
    
    # First layer
    x = Conv2D(filters=f1, kernel_size=1, strides=1, padding='valid')(x)
    x = BatchNormalization(axis=3)(x)
    x = ReLU()(x)

    # Second layer
    x = Conv2D(filters=f2, kernel_size=ks, strides=1, padding='same')(x)
    x = BatchNormalization(axis=3)(x)
    x = ReLU()(x)

    # Third layer
    x = Conv2D(filters=f3, kernel_size=1, strides=1, padding='valid')(x)
    x = BatchNormalization(axis=3)(x)
 
    x = Add()([x, x_shortcut])
    x = ReLU()(x)
    
    return x
```

### Convolutional Block

<img src="images/Convolutional-Block.png" style="width=45px;height=45"/>

```python
def convolutional_block(x, ks, filters):

    f1, f2, f3 = filters
    x_shortcut = x

    # First layer 
    x = Conv2D(filters=f1, kernel_size=1, strides=2)(x)
    x = BatchNormalization(axis=3)(x)
    x = ReLU()(x)
    
    # Second layer
    x = Conv2D(filters=f2, kernel_size=ks, strides=1, padding='same')(x)
    x = BatchNormalization(axis=3)(x)
    x = ReLU()(x)

    # Third layer
    x = Conv2D(filters=f3, kernel_size=1, strides=1, padding='valid')(x)
    x = BatchNormalization(axis=3)(x)

    # Configuring Shortcut Path
    x_shortcut = Conv2D(filters=f3, kernel_size=1, strides=2, padding='valid')(x_shortcut)
    x_shortcut = BatchNormalization(axis=3)(x_shortcut)

    # Adding shortcut value
    x = Add()([x, x_shortcut])
    x = ReLU()(x)   
    
    return x
```

### Buliding ResNet-50

```python
def resnet50(input_shape=(224, 224, 3), n_classes=1000, weights=None):

    x_input = Input(input_shape)
    x = ZeroPadding2D(padding=3)(x_input)

    # Block 1
    x = Conv2D(filters=64, kernel_size=7, strides=2)(x) 
    x = BatchNormalization(axis=3)(x) 
    x = ReLU()(x)
    x = MaxPool2D(pool_size=3, strides=2)(x)

    # Block 2
    x = convolutional_block(x, 3, [64, 64, 256])
    x = identity_block(x, 3, [64, 64, 256])   
    x = identity_block(x, 3, [64, 64, 256])

    # Block 3 
    x = convolutional_block(x, 3, [128, 128, 512])
    x = identity_block(x, 3, [128, 128, 512])
    x = identity_block(x, 3, [128, 128, 512])
    x = identity_block(x, 3, [128, 128, 512])

    # Block 4 
    x = convolutional_block(x, 3, [256, 256, 1024])
    x = identity_block(x, 3, [256, 256, 1024])
    x = identity_block(x, 3, [256, 256, 1024])
    x = identity_block(x, 3, [256, 256, 1024])
    x = identity_block(x, 3, [256, 256, 1024])
    x = identity_block(x, 3, [256, 256, 1024])

    # Block 5 
    x = convolutional_block(x, 3, [512, 512, 2048])
    x = identity_block(x, 3, [512, 512, 2048])
    x = identity_block(x, 3, [512, 512, 2048])

    # Avg Pool 
    x = AveragePooling2D(pool_size=2)(x)

    # Output layer
    x = Flatten()(x)
    output = Dense(n_classes, activation='softmax', kernel_initializer=glorot_uniform(seed=0))(x)
    
    model = Model(inputs=x_input, outputs=output)
    
    if weights is not None:
        model.load_weights(weights, by_name=True)

    return model
```

```python
model = resnet50()
plot_model(model, to_file='ResNet50Model.png', show_shapes=True)
```
